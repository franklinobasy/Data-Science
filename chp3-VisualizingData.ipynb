```python
#plot data
x = [1, 2, 3, 4]
y = [5, 4, 3, 2]

#Create a new figure
figure("Charts")

#divide subplots into 2 x 3 grids
#and select grid 1
subplot(231)
plot(x, y)

#select grid 2
subplot(232)
bar(x, y)

#select grid 3
#horizontal bar-charts
subplot(233)
barh(x, y)

#select grid 4
#Create stacked bar charts
subplot(234)
bar(x, y)
#----we need more data for stacked bar charts
y1 = [7, 8, 5, 3]
bar(x, y1, bottom=y, color = 'r')

#select grid 5
subplot(235)
boxplot(x)

#select grid 6
subplot(236)
scatter(x, y)


# show figure
show()
```

```python
''' Understanding Box plots'''

from pylab import *

figure()
dataset = [113, 115, 119, 121, 124,
            124, 125, 126, 126, 126,
            127, 127, 128, 129, 130,
            130, 131, 132, 133, 136]

subplot(121)
boxplot(dataset, vert = False)

subplot(122)
hist(dataset)

show()
```
